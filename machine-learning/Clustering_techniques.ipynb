```python
from sklearn.metrics import silhouette_score, silhouette_samples
from sklearn.cluster import KMeans, dbscan
from sklearn.datasets import make_blobs
import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
```

```python
# generating samples, by using  make_blobs
# 1 distinct cluster, and 3 close clusters
X, y = make_blobs(n_samples=500,
                  n_features=2,
                  centers=4,
                  cluster_std=1,
                  center_box=(-10.0, 10.0),
                  shuffle=True,
                  random_state=1)
```

```python
# Elbow method for finding best k , in range 1 -> 11
inertia_score = []
for k in range(1,12):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(X)
    inertia_score.append(kmeans.inertia_)

plt.plot(range(1,12),inertia_score)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Choosing number of cluster as 3 or 4 gives us the most abrupt decrease in WCSS
kmeans = KMeans(n_clusters=4, random_state=0)
y_pred = kmeans.fit_predict(X)
print(y_pred)
```
